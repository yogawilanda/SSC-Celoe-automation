```python
import pandas as pd

filename = 'aset/asetEdomGanjil2223.xlsx'
xl = pd.ExcelFile(filename)

# Get sheet names
sheet_names = xl.sheet_names

# Create a writer for the merged data
writer = pd.ExcelWriter('aset/mergedData.xlsx', engine='xlsxwriter')

# Create an empty DataFrame to store combined data
combined_data = pd.DataFrame()

# Iterate over sheet names
for sheet_name in sheet_names:
    # Read each sheet
    df = xl.parse(sheet_name, parse_dates=True)
    
    # Concatenate the data to the combined_data DataFrame
    combined_data = pd.concat([combined_data, df], ignore_index=True)

# Write the combined data to a new Excel file
combined_data.to_excel(writer, index=False, sheet_name='CombinedData')

# Save the Excel file
writer.close()
```

```python
import pandas as pd

def insert_empty_column_with_sheet_name(filename, sheet_name):
    """
    Inserts an empty column with the sheet name in the next available empty column.
    If no empty columns are found, handles the scenario gracefully.

    Args:
        filename (str): The path to the Excel file.
        sheet_name (str): The name of the sheet to process.
    """

    # Open the Excel file in append mode with the appropriate engine
    engine = "openpyxl"  # Use openpyxl for .xlsx files
    writer = pd.ExcelWriter(filename, engine=engine, mode="a")

    # Load the specific sheet into a DataFrame
    df = pd.read_excel(filename, sheet_name=sheet_name, engine=engine)

    # Find the index of the next empty column, handling potential absence of empty columns
    try:
        next_empty_column = next(col for col in df.columns if df[col].isnull().all())
    except StopIteration:
        print(f"No empty columns found in sheet '{sheet_name}'. Skipping...")
        return  # Exit the function if no empty columns are found

    # Insert the new empty column with the sheet name
    df.insert(next_empty_column, sheet_name, pd.NA)

    # Write the modified DataFrame back to the Excel file
    df.to_excel(writer, sheet_name=sheet_name, index=False)

    # Close the writer to save changes
    writer.close()

# Example usage:
filename = 'aset/asetEdomGanjil2223.xlsx'
for sheet_name in pd.ExcelFile(filename).sheet_names:  # Iterate through all sheets
    insert_empty_column_with_sheet_name(filename, sheet_name)
```
